# CycleAnalyzer usage demonstration

```python
from chronobiology.chronobiology import DBQuery, CycleAnalyzer, generate_data
```

```python
# Generate random data to use as input.
data = generate_data()
timestamps = data['time']
values = data['value']
nights = data['is_night']
print(f'{len(timestamps)} measurements generated.')
```

```python
# Create cycle analyzer.
ca = CycleAnalyzer(timestamps, values, nights, descr='Sample series')
print(f'Time range is {ca.timestamps.min()} - {ca.timestamps.max()}')
```

## Actogram

```python
ca.plot_actogram()
```

## Periodogram

```python
ca.plot_periodogram()
```
